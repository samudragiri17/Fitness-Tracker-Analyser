```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

users = [101, 102, 103, 104, 105]
dates = [datetime(2024, 6, 1) + timedelta(days=i) for i in range(30)]
data = []

for date in dates:
    for user in users:
        steps = np.random.randint(3000, 15000)
        calories = np.random.randint(150, 500)
        heart_rate = np.random.randint(60, 110)
        sleep = round(np.random.normal(6.5, 1), 1)
        data.append([date.strftime('%Y-%m-%d'), user, steps, calories, heart_rate, sleep])

df = pd.DataFrame(data, columns=["Date", "User_ID", "Steps", "Calories_Burned", "Heart_Rate", "Sleep_Hours"])
df.to_csv("fitness_data.csv", index=False)

df['Date'] = pd.to_datetime(df['Date'])  # Parse date column
df.head()

print(df.isnull().sum())  # Check for missing values
# Optional: df = df.dropna() or df.fillna(method='ffill')

# Sleep Quality Classification
df['Sleep_Quality'] = np.where(df['Sleep_Hours'] >= 7, 'Good',
                        np.where(df['Sleep_Hours'] >= 6, 'Average', 'Poor'))

# Add Week column
df['Week'] = df['Date'].dt.isocalendar().week
print(df['Sleep_Quality'])
print(df['Week'])

# Average steps per user
print(df.groupby('User_ID')['Steps'].mean())

# Weekly summary
weekly_summary = df.groupby(['User_ID', 'Week'])[['Steps', 'Calories_Burned']].sum()
print(weekly_summary.head())

# Line Plot: Steps Over Time
for uid in df['User_ID'].unique():
    plt.plot(df[df['User_ID'] == uid]['Date'], df[df['User_ID'] == uid]['Steps'], label=f'User {uid}')
plt.title('Steps Over 30 Days')
plt.xlabel('Date')
plt.ylabel('Steps')
plt.legend()
plt.grid()
plt.show()

# Bar Plot: Average Calories Burned
df.groupby('User_ID')['Calories_Burned'].mean().plot(kind='bar', color='orange')
plt.title('Avg Calories per User')
plt.ylabel('Calories')
plt.grid()
plt.show()

# Pie Chart: Sleep Quality
df['Sleep_Quality'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sleep Quality Distribution')
plt.ylabel('')
plt.show()

# Scatter Plot: Sleep vs Calories
plt.scatter(df['Sleep_Hours'], df['Calories_Burned'], c='green')
plt.xlabel('Sleep Hours')
plt.ylabel('Calories Burned')
plt.title('Sleep vs Calories Burned')
plt.grid()
plt.show()

df.to_csv("cleaned_fitness_data.csv", index=False)
```
